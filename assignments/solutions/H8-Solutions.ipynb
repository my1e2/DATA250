# Use Python and show your code.  Find four vectors in four dimensional space, with integer coordinates, whose dot products satisfy


```python
import numpy as np

# vectors
for a1, a2, a3, a4 in [(1, 1, 2, 2), (1, 2, 1, 2), (2, 1, 1, 2), (3, 1, 0, 0)]:
    u1 = np.array([a1, a2, a3, a4])
    if np.dot(u1, u1) == 10:
        break

# u2 should be orthogonal to u4 and have length 2
u2 = np.array([2, 0, 0, 0])

# u3 needs to be length 4 and satisfy the dot products
u3 = np.array([0, -4, 0, 0])

# u4 needs to satisfy all remaining conditions
u4 = np.array([-1, 1, 0, -1])

print("Vectors found:")
print(f"u1 = {u1}")
print(f"u2 = {u2}")
print(f"u3 = {u3}")
print(f"u4 = {u4}")

print("\nChecking all conditions:")
conditions = [
    ("u1·u1 = 10", np.dot(u1, u1), 10),
    ("u1·u2 = 4", np.dot(u1, u2), 4),
    ("u1·u3 = 4", np.dot(u1, u3), 4),
    ("u1·u4 = -3", np.dot(u1, u4), -3),
    ("u2·u2 = 4", np.dot(u2, u2), 4),
    ("u2·u3 = -4", np.dot(u2, u3), -4),
    ("u2·u4 = 0", np.dot(u2, u4), 0),
    ("u3·u3 = 16", np.dot(u3, u3), 16),
    ("u3·u4 = -6", np.dot(u3, u4), -6)
]

for name, actual, expected in conditions:
    status = ", Works" if actual == expected else ", Doesn't Work"
    print(f"{name}: {actual}{status}")
```

# Suppose that we have the following dot products.

dot(u, u) = 18

dot(u, v) = 6

dot(v, v) = 4

Calculate dot(u + v, u - 2v)

```python
# Given dot products
u_dot_u = 18
u_dot_v = 6
v_dot_v = 4

# Compute (u + v) · (u - 2v)
result = u_dot_u - u_dot_v - 2 * v_dot_v

print("dot(u + v, u - 2v) =", result)
```
