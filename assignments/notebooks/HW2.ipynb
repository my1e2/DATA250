# Question X: Code the following in Python.

Define q[j] = minimum (j+1, 6-j)/12

Let p0 be a list of length 6 with random values between 0 and 1.

Define p = p0 / the total sum of p0 so that p is a probability vector, as is q.

Calculate -sum(p(j)lnq(j)) from j=0 to 5



```python
import numpy as np

# defining q[j] = min(j+1, 6-j)/12 within the same range as p
q = [min(j+1, 6-j)/12 for j in range(6)]

# creating p0 with random values between 0 and 1
p0 = np.random.random(6)

# making p a probability vector 
p = p0 / np.sum(p0)

# calculating -sum p(j) * ln(q(j)) from indices 0 to 5
value = -sum(p[j] * np.log(q[j]) for j in range(6))
print("Sum:", value)
```

# Question Y: Code the following in Python.

Define f(m) = sum(sqrt(j)) from j=0 to m

Make a scatter plot of f(m) from m=0 to 100

```python
import matplotlib.pyplot as plt
import numpy as np

# function to define sum f(m) and its index

def f(m):
    
    return sum(np.floor(np.sqrt(j)) for j in range(m + 1))

# generating values from 0 to 100
v1 = list(range(101))
v2 = [f(m) for m in v1]

# scatter plot functionality
plt.figure(figsize=(8, 5))
plt.scatter(v1, v2, alpha=0.5, color='red')
plt.title("Scatter plot of f(m) = sum floor(sqrt(j)) from j=0 to m")
plt.xlabel('m')
plt.ylabel('f(m)')
plt.grid(True, alpha=0.5)
plt.show()
```

# Question Z: Code the following in Python.

Find one instance where the following logical expression is true.

!(Nand[ !(a || c || b), Nand[b, e]])

which in Python code is (assuming you have defined nand)

not( nand( not(a or c or b), nand(b, e)))

```python
# defining nand functionality

def nand(a, b):

    return not (a and b)

# initializing as false and setting combinations 

found = False
for a in [False, True]:
    for b in [False, True]:
        for c in [False, True]:
            for e in [False, True]:

                # cycling through and checking for expression

                expression = not(nand(not(a or c or b), nand(b, e)))
                
                if expression:
                    print(f"True instance: a={a}, b={b}, c={c}, e={e}")
                    found = True
                    break
            if found:
                break
        if found:
            break
    if found:
        break

if not found:
    print("No true instance")
```
