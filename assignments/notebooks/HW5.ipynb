Let A ={a,b,c}, B={x,y}, and C = {0,1}. 

Find A x B x C by listing its elements using a Python program.

```python
# sets
A = {'a', 'b', 'c'}
B = {'x', 'y'}  
C = {'0', '1'}

# cartesian product A × B × C
cartesian_product = []

for a in A:
    for b in B:
        for c in C:
            cartesian_product.append((a, b, c))

print("A × B × C =", cartesian_product)
print("Number of elements:", len(cartesian_product))
```
