# Question X:

Write a function in Python to compute the n-th Fibonacci number.

# Question Y:

Write a function in Python that computes the n-th number of the recursive sequence

f(n) = f(f(n-1)) + f(n - f(n-1)) with f(1) = f(2) = 1

Compute f(100) and make a scatter plot of the values up to n=1000

```python
# Question X
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

print("Fibonacci numbers:")
for i in range(10):
    print(f"f({i}) = {fib(i)}")

print(f"f(10) = {fib(10)}")

# Question Y
def f(n_max):
    seq = [0, 1, 1]  # f(0), f(1), f(2)
    for i in range(3, n_max + 1):
        seq.append(seq[seq[i-1]] + seq[i - seq[i-1]])
    return seq

# f(100)
result = f(100)
print(f"f(100) = {result[100]}")

# scatter plot
import matplotlib.pyplot as plt

values = f(1000)
plt.scatter(range(1, 1001), values[1:], s=1)
plt.show()
```
