Question X: Using Python, compute the values up to n = 20 for fn = fn-1 + fn-2 + fn-3 with the initial condition f0 = 1, f1 = 2, f2 = 3

```python
# Question X:

# function to set initial conditions

def value_computer(n):
    if n == 0:
        return 1
    elif n == 1:
        return 2
    elif n == 2:
        return 3
    
    # initializing f/sequence at each value/length
    f = [1, 2, 3]
    
    # computing values up until n (which is 20)
    for value in range(3, n + 1):
        values = f[value-1] + f[value-2] + f[value-3]
        f.append(values)
    
    return f

# n = 20 result
n = 20

# calling function to assign what to do with values 0, 1, and 2 

result = value_computer(n)

print("Printing values up to n = 20:")
for iterable, value in enumerate(result):
    print(f"f{iterable} = {value}")
```
