Use numpy and matplotlib for the following.

a)  Create a list or array of 50 random numbers between 0 and 1, and set the variable x to be this list.

b)  Create another list of 50 random numbers between 0 and 1, and set the variable y to be this list.

c)  Calculate x+y and x-y

d)  Calculate x dot y

e)  Take the elements 10 to 20 of x and set z to be this list.  Make a plot of z.  

```python
import numpy as np
import matplotlib.pyplot as plt

# a) 
x = np.random.random(50)

# b) 
y = np.random.random(50)

# c) 

x_plus_y = x + y
x_minus_y = x - y

# d) 
x_dot_y = np.dot(x, y)

# e) 
z = x[10:21]
plt.plot(z)
plt.title('Plot of z')
plt.xlabel('Indices')
plt.ylabel('Values')
plt.grid(True)
plt.show()

print(f"x dot y = {x_dot_y}")
```
